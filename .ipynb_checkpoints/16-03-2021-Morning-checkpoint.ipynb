# Preprocessing 

## Robust scaler
-  If your data set contains more number of outliers, then robust scaling is best technique for scaling
-  

### Robust scaler formula

# Xi - Q1(X)
# -------------
# Q3(X)-Q1(X)

###Inter Quartile Range(IQR)
# Q1 ----> 25%(Lower quartile)
# Q2 ----> 50%(Medium quartile)
# Q3----> 75%(Upper quartile)

```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
```

```python
data1.head(21)
```

```python
data1.columns
```

```python
data1.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
data1.head()
```

# Apply robust scaler model

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
rs
```

```python
rs2 = rs.fit_transform(data1)
rs2
```

```python
# to convert the output into dataframes
rs2= pd.DataFrame(rs2, columns=data1.columns)
rs2.head()
```
